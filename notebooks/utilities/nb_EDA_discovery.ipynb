# Exploratory Data Analysis Discovery

This notebook performs exploratory data analysis (EDA) on various raw datasets related to the project.
The notebook provides an overview of each dataset to facilitate further analysis and data discovery.

```python
%run ../nb_setup
```

### 1. accounts (csv)

```python
display(spark.read.format("csv")
        .option("header", "true")
        .load(f"{paths['raw_data']}/accounts")
        )
```

### 2. branches (csv)

```python
display(spark.read.format("csv")
        .option("header", "true")
        .load(f"{paths['raw_data']}/branches")
        )
```

### 3. exchange_rates (csv)

```python
display(spark.read.format("csv")
        .option("header", "true")
        .load(f"{paths['raw_data']}/exchange_rates"))
```

### 4. transactions (csv)

```python
display(spark.read.format("csv")
        .option("header", "true")
        .load(f"{paths['raw_data']}/transactions"))
```

### 5. customers (json)

```python
display(spark.read.format("json")
        .option("multiLine", "true") 
        .load(f"{paths['raw_data']}/customers"))
```

### 6. credit_scores (parquet)

```python
display(spark.read.format("parquet")
        .load(f"{paths['raw_data']}/credit_scores")
        )
```

### 7. audit_logs (avro)

```python
display(spark.read.format("avro").load(f"{paths['raw_data']}/audit_logs"))
```
