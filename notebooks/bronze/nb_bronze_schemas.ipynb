# Schema Registry: Bronze Layer

This notebook defines the schema for the initial ingestion. 

Every field is defined as `StringType()`. 
This ensures that the ingestion pipeline never fails due to source data type mismatches. 
Data quality and type enforcement are deferred to the Silver layer.

```python
from pyspark.sql.types import StructType, StructField, StringType
```

```python
accounts_schema = StructType([
    StructField("account_id", StringType(), True),
    StructField("cust_id", StringType(), True),
    StructField("account_type", StringType(), True),
    StructField("balance", StringType(), True),
    StructField("open_date", StringType(), True)
])
```

```python
audit_logs_schema = StructType([
     StructField("event_id", StringType(), True),
     StructField("user_id", StringType(), True),
     StructField("action", StringType(), True),
     StructField("ip_address", StringType(), True),
     StructField("device_info", StringType(), True)
 ])
```

```python
barnches_schema = StructType([
    StructField("branch_id", StringType(), True),
    StructField("name", StringType(), True),
    StructField("city", StringType(), True)
])
```

```python
credit_scores_schema = StructType([
     StructField("cust_id", StringType(), True),
     StructField("credit_score", StringType(), True),
     StructField("risk_level", StringType(), True)
 ])
```

```python
customers_schema = StructType([
     StructField("address", StringType(), True),
     StructField("cust_id", StringType(), True),
     StructField("email", StringType(), True),
     StructField("full_name", StringType(), True),
     StructField("phone", StringType(), True),
     StructField("updated_at", StringType(), True)
 ])
```

```python
exchange_rates_schema = StructType([
    StructField("currency", StringType(), True),
    StructField("rate_to_usd", StringType(), True),
    StructField("date", StringType(), True)
])
```

```python
transactions_schema = StructType([
     StructField("tx_id", StringType(), True),
     StructField("src_account", StringType(), True),
     StructField("dst_account", StringType(), True),
     StructField("amount", StringType(), True),
     StructField("tx_type", StringType(), True),
     StructField("timestamp", StringType(), True)
 ])
```
